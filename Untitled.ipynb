```python
train_data_2 = train_data_0.iloc[0:len(train_data_1)]
```

```python
train_data = pd.concat([train_data_1, train_data_2], axis=0)
```

```python
train_class_x = train_data.iloc[:, :-2]
train_class_y = train_data.iloc[:,-1]
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(train_class_x, train_class_y, test_size=0.25, random_state=33)
```

```python
cmf = ClassifierModelFactory()
model = cmf.create_model(ClassifierModelFactory.MODEL_XGBOOST)
model.fit(X_train, X_valid, y_train, y_valid)
predict = model.predict(X_valid)
print(predict)
```

```python
ss = lambda num : 1 if num >0.7 else 0
```

```python
prediction = [ss(p) for p in predict]
```

```python
valid = pd.DataFrame(y_valid.values, columns=['valid'])
```

```python
valid['predict'] = prediction
```

```python
valid[(valid['valid'] == 0).values & (valid['predict'] == 1).values]
```

```python
accuracy_score(y_valid, prediction)
```
